```python
# Task 1: Convert images to 8-bit unsigned RGB encoding

def convert_to_RGB(image):
    image = image[:, :, :3]
    image = image * 255
    image = np.uint8(np.clip(image.round(), 0, 255))
    return image
```

```python
img2 = convert_to_RGB(img2)
```

```python
plt.imshow(img2)
```
