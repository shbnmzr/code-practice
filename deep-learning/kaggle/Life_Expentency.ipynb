```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# 1 - Data Import and Analysis

```python
life_exp_data = pd.read_csv('life-expentency.csv', index_col='ISO3')
life_exp_data
```

```python
life_exp_years = life_exp_data.loc[:, 'Life Expectancy at Birth (1990)':'Life Expectancy at Birth (2021)']
life_exp_years
```

```python
min_data = life_exp_years.min()
max_data = life_exp_years.max()
mean_data = life_exp_years.mean()
```

```python
plt.figure(figsize=(16, 8))
sns.lineplot(min_data, label='Min')
sns.lineplot(max_data, label='Max')
sns.lineplot(mean_data, label='Mean')

plt.legend()

plt.show()
```

```python
continental_groups = life_exp_data.groupby(by='Continent')
```

```python
continental_groups.describe()
```

```python
def get_group_data(key: str):
    return continental_groups.get_group(key).loc[:, 'Life Expectancy at Birth (1990)': 'Life Expectancy at Birth (2021)']
```

```python
oceania = get_group_data('Oceania')
asia = get_group_data('Asia')
africa = get_group_data('Africa')
europe = get_group_data('Europe')
america = get_group_data('America')
```

```python
plt.figure(figsize=(16, 8))
sns.lineplot(oceania.mean(), label='Oceania')
sns.lineplot(asia.mean(), label='Asia')
sns.lineplot(america.mean(), label='America')
sns.lineplot(europe.mean(), label='Europe')
sns.lineplot(africa.mean(), label='Africa')

plt.show()
```

```python
plt.figure(figsize=(16, 8))
sns.lineplot(oceania.min(), label='Oceania')
sns.lineplot(asia.min(), label='Asia')
sns.lineplot(america.min(), label='America')
sns.lineplot(europe.min(), label='Europe')
sns.lineplot(africa.min(), label='Africa')

plt.show()
```

```python
plt.figure(figsize=(16, 8))
sns.lineplot(oceania.max(), label='Oceania')
sns.lineplot(asia.max(), label='Asia')
sns.lineplot(america.max(), label='America')
sns.lineplot(europe.max(), label='Europe')
sns.lineplot(africa.max(), label='Africa')

plt.show()
```

```python
plt.figure(figsize=(16, 8))
sns.lineplot(oceania.median(), label='Oceania')
sns.lineplot(asia.median(), label='Asia')
sns.lineplot(america.median(), label='America')
sns.lineplot(europe.median(), label='Europe')
sns.lineplot(africa.median(), label='Africa')

plt.show()
```
